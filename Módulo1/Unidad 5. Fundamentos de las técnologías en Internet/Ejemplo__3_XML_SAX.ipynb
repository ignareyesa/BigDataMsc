```python
import xml.sax

class prueba (xml.sax.ContentHandler):
    def __init__(self):
        self.Book = ""
        self.Author = ""
        self.Title = ""
        self.Genre = ""
        self.Price = ""
        self.PublishDate = ""
        self.Description = ""        
        
    def startElement(self,tag, atribute):
        self.Book = tag
        if tag == "Book" :
            print("***Book***")
            ids = atribute["ids"]
            print("id: ", ids)
        
    def endElement(self,tag):
        if self.Book == "Author":
            print("Author: ", self.Author)
        if self.Book == "Title":
            print("Title: ", self.Title)
        if self.Book == "Genre":
            print("Genre: ", self.Genre)
        if self.Book == "Price":
            print("Price: ", self.Price)
        if self.Book == "PublishDate":
            print("PublishDate: ", self.PublishDate)
        if self.Book == "Description":
            print("Description: ", self.Description)
        self.Book = ""
    
    def characters(self,content):
        if self.Book == "Author":
            self.Author = content
        if self.Book == "Title":
            self.Title = content
        if self.Book == "Genre":
            self.Genre = content
        if self.Book == "Price":
            self.Price = content
        if self.Book == "PublishDate":
            self.PublishDate = content
        if self.Book == "Description":
            self.Description = content
        

if ( __name__ == "__main__"):
    # Crear un XMLReader
    parser=xml.sax.make_parser()
    # Deshabilitar namespaces
    parser.setFeature(xml.sax.handler.feature_namespaces,0)
    #Sobre escribir el default default ContextHandler
    Handler=prueba()
    
    parser.setContentHandler(Handler)
    parser.parse("books.xml") 
```
