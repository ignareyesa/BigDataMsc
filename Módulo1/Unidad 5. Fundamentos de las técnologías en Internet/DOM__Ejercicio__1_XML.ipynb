```python
import urllib
import xml.dom
from urllib import request
from xml.dom.minidom import parse

#local
arbol_dom = parse("Datos/purchaseorder_extended.xml")
#online
#url = "https://raw.githubusercontent.com/ignareyesa/BigDataMsc/master/M%C3%B3dulo1/Unidad%205.%20Fundamentos%20de%20las%20t%C3%A9cnolog%C3%ADas%20en%20Internet/Datos/purchaseorder.xml"
#arbol_dom = parse(request.urlopen(url))


purchases = arbol_dom.getElementsByTagName("PurchaseOrder")
for purchase in purchases:
    
    print("Order Number: ", purchase.getAttribute("PurchaseOrderNumber"))
    print("Order Date: ", purchase.getAttribute("OrderDate"))
    print("---")
    addresses = purchase.getElementsByTagName("Address")
    for address in addresses:
        print("Type: ", address.getAttribute("Type"))
        print("Name: ", address.getElementsByTagName("Name")[0].firstChild.data)
        print("Street: ", address.getElementsByTagName("Street")[0].firstChild.data)
        print("City: ", address.getElementsByTagName("City")[0].firstChild.data)
        print("State: ", address.getElementsByTagName("State")[0].firstChild.data)
        print("Zip: ", address.getElementsByTagName("Zip")[0].firstChild.data)
        print("Country: ", address.getElementsByTagName("Country")[0].firstChild.data)
        print("---")
    print("Delivery Notes: ", purchase.getElementsByTagName("DeliveryNotes")[0].firstChild.data)
    print("---")
    items = purchase.getElementsByTagName("Items")
    for item in items:
        its = item.getElementsByTagName("Item")
        for it in its:
            print("Part Number: ", it.getAttribute("PartNumber"))
            print("Product Name: ", it.getElementsByTagName("ProductName")[0].firstChild.data)
            print("Quantity: ", it.getElementsByTagName("Quantity")[0].firstChild.data)
            print("USPrice: ", it.getElementsByTagName("USPrice")[0].firstChild.data)
            print("Comment: ", it.getElementsByTagName("Comment")[0].firstChild.data)
            print("---")
    print("---------------------------------------------------------------")
        
```
